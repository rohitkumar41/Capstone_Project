```python
# Now in the price column the price has two unit one is in lac and the other is in crore we will try to convert the lac  into crore so that it can be only one unit 
```

```python
def treat_price(x):
    if type(x) == float:
        return x
    else:
        if x[1] == 'Lac':
            return round(float(x[0])/100,2)
        else:
            return round(float(x[0]),2)
```

```python
df.head()
```

```python
df['price'] = df['price'].str.split(' ').apply(treat_price)
```
